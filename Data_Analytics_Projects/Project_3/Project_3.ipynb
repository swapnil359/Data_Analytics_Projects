## Replace Salary Values With 0 and 1

```python
data['income'].unique()
```

```python
data['income'].value_counts()
```

```python
sns.countplot(x='income',data=data)
```

```python
def income_data(inc):
    if inc == '<=50K':
        return 0
    elif inc == '>50K':
        return 1
    else:
        return None  

    
```

```python
data['encoded_salary']=data['income'].apply(income_data)
```

```python
data.head()
```

### 2nd method

```python
data.replace(to_replace=['<=50K', '>50K'],value=[0,1],inplace = True)
```
